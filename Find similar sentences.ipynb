```python
import numpy as np
from scipy.spatial.distance import cosine
import re
```

```python
with open('text2.txt', 'r') as file:
    sentences = file.readlines()

all_words = []
```

```python
for item in sentences:
    all_words += re.split('[^a-z]', item.lower())
```

```python
all_words = list(filter(None, all_words))
```

```python
words_dict = {}
index = 0
```

```python
for word in all_words:
    if word in words_dict:
        continue
    else:
        words_dict[word] = index
        index += 1
```

```python
text_matrix = np.zeros((len(sentences), len(words_dict)), dtype=int)
```

```python
def fill_the_vector(line):
    """identifies each single word in sentence, counts it in the sentence and fills respective feature in the vector"""
    
    #импортируем словарь со всеми словами текста
    global words_dict
    
    #создаем вектор длиной, равной длине словаря
    vector = np.zeros(len(words_dict))
    
    #разбираем входное предложение на отдельные слова
    line = re.split('[^a-z]', line.lower())
    
    #для каждого слова в словаре проверяем, входит ли оно в предложение. Если да, увеличваем на 1 счетчик соответствующего признака
    for word in words_dict:
        for single_word in line:
            if word == single_word:
                vector[words_dict.get(word)] += 1
            else:
                continue
    
    return(vector)
```

```python
for index, single_sentence in enumerate(sentences):
    text_matrix[index] = fill_the_vector(single_sentence)
```

```python
distances = np.zeros(len(text_matrix))
```

```python
for item, single_sententence_descriptor in enumerate(text_matrix):
    distances[item] = cosine(text_matrix[0], single_sententence_descriptor)
```

```python
distances
```

```python
for item in np.argsort(distances):
    print(sentences[item])
```
